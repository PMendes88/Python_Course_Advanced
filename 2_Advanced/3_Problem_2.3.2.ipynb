# Random Sample of DataFrame (2.3.2) - Problem

Problem Statement: Create a new DataFrame that is a random sample of 10% of the original DataFrame. Display the first 5 rows of the sampled DataFrame.

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datadataset 
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_random = df.copy()

df_first_five = df_random.sample(10, random_state=42)

df_first_five.head(5)
```

```python
# Luke's Solution

df_sampled = df.sample(frac=0.1)
df_sampled.head(5)
```
