# Job Schedule Distribution (2.14.2) - Problem

Problem Statement: Create a pie chart showing the count of job postings per job schedule types (for only jobs that are either 'Full-time' and 'Part-time') in the DataFrame.
Create a DataFrame to filter out these job_schedule_type called df_ft_pt.
Plot this using plt.pie(), setting labels parameter to the index, startangle to 90, and format values to display similarly to 23.1%.
Add a title 'Job Schedule Full-time vs. Part-time'.

```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_ft_pt = df[df['job_schedule_type'].isin(['Full-time', 'Part-time'])]

schedule_counts = df_ft_pt['job_schedule_type'].value_counts()


plt.pie(schedule_counts, labels=schedule_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Job Schedule: Full-time vs. Part-time')
plt.show


```
