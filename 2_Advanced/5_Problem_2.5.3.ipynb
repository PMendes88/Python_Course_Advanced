#Sort Index (2.5.3) - Problem

Problem Statement: Create a pivot table to calculate the median salary_year_avg for each company_name and sort the index in descending order. Display the first 5 rows.

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datadataset 
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_pivot = df.pivot_table(values='salary_year_avg', index='company_name', aggfunc='median')

df_pivot
```

```python
df_pivot.sort_index(inplace=True, ascending=False)

df_pivot
```
