# Pivot Table Reset & Set Index (2.5.4) - Problem

Problem Statement: Create a pivot table to calculate the median salary_year_avg for each job_title_short, reset the index of the resulting pivot table, then set the index to salary_year_avg. Display the first 5 rows.

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datadataset 
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_pivot = df.pivot_table(index='job_title_short', values='salary_year_avg', aggfunc='median')

df_pivot
```

```python
df_pivot.reset_index(inplace=True)

df_pivot
```

```python
df_pivot.set_index('salary_year_avg', inplace=True)

df_pivot.head()
```
