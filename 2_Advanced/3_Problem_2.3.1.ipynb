# Drop Missing Values (2.3.1) - Problem

Problem Statement: Create a copy of the DataFrame and drop rows where the salary_year_avg column has missing values. Display the first 5 rows of the cleaned DataFrame.

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datadataset 
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_original = df
```

```python
df_altered = df_original.copy()

df_altered.loc[:5, 'salary_year_avg']

df_cleaned = df_altered.dropna(subset=['salary_year_avg'])

df_cleaned.head(5)
```
